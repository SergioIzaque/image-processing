aqui guarda a imagem e mostra o nome dela.


```python
from matplotlib import pyplot as plt
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

aqui apresenta a imagem, convertendo ela de bgr para rgb.

```python
x = img.shape [0]
y = img.shape [1]
z = img.shape [2]

for r in range (0,4):
  novox = x//2
  novoy = y//2
  img1 = np.zeros((novox,novoy,z), dtype=np.uint8)
  for i in range (0,novox):
    for j in range (0,novoy):
      for k in range (0,z):
        img1[i,j,k] = (img[2*i,2*j,k]*1 + img[2*i+1,2*j,k]*1 + img[2*i,2*j,k]*1 + img[2*i+1,2*j+1,k]*1)/4

  plt.imshow(img1)
  plt.figure()
  x = novox
  y = novoy
  img = img1
```

nessa parte, primeiramente ele divide o x para altura da imagem
o y para a largura
e o z para as cores (3 pois √© RGB)

o for r in range √© para diminuir 4 vezes a imagem

depois √© criada uma imagem na escala menor.

abaixo √© onde ocorre a redu√ß√£o de resolu√ß√£o.
cada 4 pixels da imagem original √© reduzido a 1 pixel.
os valores s√£o somados e dividos por 4, fazendo uma m√©dia simples.

a nova imagem √© apresentada.

e as dimens√µes s√£o atualizadas.

```python
for r in range (0,4):
  novox = x*2
  novoy = y*2
  img1 = np.zeros((novox,novoy,z), dtype=np.uint8)
  for i in range (x):
    for j in range (y):
      for k in range (z):
        img1[2*i,2*j,k] = img[i,j,k]
        img1[2*i+1,2*j,k] = img[i,j,k]
        img1[2*i,2*j+1,k] = img[i,j,k]
        img1[2*i+1,2*j+1,k] = img[i,j,k]

  plt.imshow(img1)
  plt.figure()
  x = novox
  y = novoy
  img = img1
```

nessa outra parte, ele faz o processo inverso, aumentando a imagem.

o for r in range vai repetir o processo 4 vezes.

depois √© feita a cria√ß√£o de uma nova imagem de tamanho aumentado.

abaixo, cada pixel da imagem √© copiado para quatro posi√ß√µes adjacentes.

depois exibe a imagem.
