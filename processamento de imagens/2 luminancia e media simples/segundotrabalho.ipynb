```python
def msimples (img):
  return np.mean(img, axis = 2).astype(np.uint8)
```

utilizando a função np.mean para obter a média

```python
def luminancia (img):
  return (0.299 * img[:,:,2] + 0.587 * img[:,:,1] + 0.114 * img[:,:,0]).astype(np.uint8)
```

```python
def conversao (img):
  return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

aqui está mudando para cinza utilizando o cv2

```python
caminho = "/content/drive/MyDrive/Doom_cover_art.jpg"
img = cv2.imread(caminho)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

aqui apresenta a imagem original

```python
imgsimplescinza = msimples(img)
imgluminancia = luminancia(img)
imgconversao = conversao(img)
```

aqui as váriaveis foram criadas para serem usadas pela função

```python
plt.imshow(imgsimplescinza, cmap = 'gray')
plt.title("Doom cinza")
```

aqui aparece a imagem com o cálculo da média simples, convertida para cinza.

```python
plt.imshow(imgluminancia, cmap = 'gray')
plt.title("Doom luminoso")
```

aqui aparece a imagem com o cálculo da média ponderada, convertida para cinza, com base na luminância.

```python
plt.imshow(imgconversao, cmap = 'gray')
plt.title("Doom convertido")
```
